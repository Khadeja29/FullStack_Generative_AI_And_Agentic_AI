```python
from sentence_transformers import SentenceTransformer
```

Downloading model to convert data into embedding

```python
open_source_embedding_model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
text = "machine"
```

```python
embedding = open_source_embedding_model.encode(text)  #we can get the embedding for the text using the open source model
```

```python
embedding.shape
```

```python
sentence = "How to reduce heart disease risk?"
```

```python
embedding_sentence = open_source_embedding_model.encode(sentence)
```

```python
embedding_sentence.shape
```

```python
paragraph = """
Machine learning is a field of artificial intelligence that focuses on building systems that learn from data. It is widely used in applications like recommendation systems, image recognition, and natural language processing.
"""
```

```python
embedding_paragraph = open_source_embedding_model.encode(paragraph)
```
