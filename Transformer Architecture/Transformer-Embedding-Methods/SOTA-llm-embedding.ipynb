Semantic Search

=== Cosine Similarity (Higher is better) ===
Range: -1 to 1

=== Dot Product (Higher is better) ===
Range: -∞ to +∞ (depends on vector magnitude)

=== Euclidean Distance (Lower is better) ===
Range: 0 to +∞

```python
import numpy as np
```

```python
def dot_product(a, b):
    return np.dot(a, b)
```

```python
def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
```

```python
def euclidean_distance(a, b):
    a = np.array(a)
    b = np.array(b)
    return np.linalg.norm(a - b)
```

```python
query = "How to reduce heart disease risk?"
```

```python
documents = [
    "Eating fiber reduces heart risk.",
    "Fruits and vegetables lower cardiovascular disease chances.",
    "Buying a new car improves driving comfort.",
    "Regular exercise helps heart health."
]
```

Convert query and documents into embeddings

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
google_embeddings_model = GoogleGenerativeAIEmbeddings(model="gemini-embedding-001")
```

```python
query_embedding = google_embeddings_model.embed_query(query)
```

```python
documents_embeddings = google_embeddings_model.embed_documents(documents)
```

```python
results = []
```

```python
for doc, doc_embedding in zip(documents, documents_embeddings):
    results.append({
        "query": query,
        "document": doc,
        "dot_product": dot_product(query_embedding, doc_embedding),
        "cosine_similarity": cosine_similarity(query_embedding, doc_embedding),
        "euclidean_distance": euclidean_distance(query_embedding, doc_embedding)
    })
```

```python
results
```

```python
def keyword_search(query, documents):
    query_words = query.lower().replace("?", "").split()
    results = []

    for doc in documents:
        doc_words = doc.lower().replace(".", "").split()
        doc_count = Counter(doc_words)

        score = sum(doc_count[word] for word in query_words) 
        results.append((doc, score))

    return sorted(results, key=lambda x: x[1], reverse=True)
```

```python
keyword_results = keyword_search(query, documents)
```
