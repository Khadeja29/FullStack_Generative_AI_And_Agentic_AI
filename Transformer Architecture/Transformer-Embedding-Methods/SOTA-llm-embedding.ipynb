```python
from sentence_transformers import SentenceTransformer
```

Downloading model to convert data into embedding

```python
open_source_embedding_model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
text = "machine"
```

```python
embedding = open_source_embedding_model.encode(text)  #we can get the embedding for the text using the open source model
```
